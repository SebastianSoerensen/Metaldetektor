```python
## Power/TX coil
E_tot = 14850 # Energi total
V = 5
I_arduino = 0.048 # Målt strømforbrug i Arduino
t = 100*60
I_corr = 0.004 # Korrektionsfaktor for strømforbrug i display osv. 
E_MCU = V*(I_arduino+I_corr)*t
K = 0.1 # Korrektionsfaktor. Reserverer K*100% af batteriet 
E_consume = E_tot*(1-K)-E_MCU
A_vmax = 9 # Max spænding
n=1 # Kun grundtonen
a_n = 2*(A_vmax/(n*sp.pi)*sp.sin((n*sp.pi)/2)) # Fourierrække
V_rms = a_n/sp.sqrt(2)
I_Tx = (E_consume/t)/V_rms # Strømmen gennem spolen

R_Tx = V_rms/I_Tx
rho_c = 1.68e-8 # Resistiviteten for kobber pr. meter
rt=     # Radius af kobbertråden i (skal angives i cm)
r=      # Radius af spolen, (skal indtastes i cm)
A=      # Højden på spolen (skal indtastes i cm)
N = (R_Tx*rt**2)/(2*r*rho_c) # Antallet af viklinger i spolen
L = (0.394*r**2*N**2)/(9*r+10*A)*1e-6 # L i hele Henry
f=2e3 # Frekvensen ved spolen
omega=2*sp.pi*f
Zl = sp.I*omega*L # Spoleimpedans
Zcsym = sp.symbols("Zcsym")
eq1 = sp.im(Zl)+sp.im(Zcsym) # Da kondensatoren skal gøre impedansen reel, skal summen af imaginærdelene være 0
Zc = sp.solve(eq1,Zcsym)[0]
Csym = sp.symbols("Csym")
eq2 = sp.Eq(Zc,1/(sp.I*omega*Csym))
C = sp.solve(eq2,Csym)[0]
```
