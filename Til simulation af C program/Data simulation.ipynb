```python
import numpy as np

FS = 8000
F  = 2000
N  = 64           # ADC_BLOCK_N
A  = 1000         # amplitude

PHASE_DEG = 30    # v√¶lg selv
PHASE = np.deg2rad(PHASE_DEG)

n = np.arange(N)

samples = np.round(
    A * np.sin(2*np.pi*F*n/FS + PHASE)
).astype(np.int16)

# --- C-identisk Goertzel ---
re = 0
im = 0

for i, s in enumerate(samples):
    if (i & 3) == 0:
        re += s
    elif (i & 3) == 1:
        im -= s
    elif (i & 3) == 2:
        re -= s
    else:
        im += s

re_n = re / N
im_n = im / N

amp = abs(re_n) + abs(im_n)
phase = np.degrees(np.arctan2(im_n, re_n))

print("Expected results:")
print("re:", re_n)
print("im:", im_n)
print("amp:", amp)
print("phase:", phase)

print("\nC array:")
for v in samples:
    print(f"{v},", end=" ")
```
